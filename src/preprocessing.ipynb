### 1. word count table for Pediatrics

```python
AbstractFil = open("../dat/Pediatrics.Abs.txt", 'rt')
wnl = WordNetLemmatizer()
stop_words = set(stopwords.words('english')) 
word_count_table = {}
for l in AbstractFil:
    try:
        abstract = l.split("|")[2]
    except:
        print(abstract)
        continue
    abstract = abstract.translate(str.maketrans('', '', string.punctuation)) # remove punctuations
    tokens = [token.lower() for token in word_tokenize(abstract)]
    lemmatized_words = [wnl.lemmatize(token) for token in tokens]
    lemmatized_words = [w for w in lemmatized_words if not w in stop_words] 
    for word in lemmatized_words:
        if word not in word_count_table:
            word_count_table[word] = 0
        word_count_table[word] += 1
AbstractFil.close()
```

```python
word_count_table_sort = sorted(word_count_table.items(), key=lambda x:x[1], reverse=True)
fout = open("../dat/Pediatrics.wordcount.txt", 'wt')
#kept_word = []
for word, count in word_count_table_sort:
    fout.write("%s\t%d\n"%(word, count))
```

```python
class arguments:
    def __init__(self, word, InpFil, outname, trainP, n_of_nearby_word, source):
        self.InpFil = InpFil
        self.outname = outname
        self.trainP = trainP
        self.word = word
        self.n_of_nearby_word = n_of_nearby_word
        self.source = source
        
def DoBagOfWords(word, source, n_of_nearby_word = 10):
    if source == "MSH":
        args = arguments(word, "../MSHCorpus/%s_pmids_tagged.arff"%(word), "../output/%s_MSH"%word, 0.7, 10, "MSH")
        ins = MyBagOfWords(args) 
        ins.run()
    elif source == "Pediatrics":
        args = arguments(word, "../dat/Pediatrics.Abs.txt", "../output/%s_Ped"%word, None, 10, "Ped")
        ins = MyBagOfWords(args) 
        ins.run()
    return ins
```
